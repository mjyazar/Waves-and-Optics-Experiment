```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

high_data = pd.read_csv('High Concentration Data.csv')

wavelengths = [510, 520, 680, 690]

# Extract wavelength and intensity data
wavelength_data = np.array(high_data['nm lambda'])
intensity_data = high_data.iloc[:, 1:].values

for wl in wavelengths:
    wavelength_index = np.argmin(np.abs(wavelength_data - wl))
    intensity_values = intensity_data[wavelength_index, :]
    print(f'Wavelength: {wl} nm, Index: {wavelength_index}, Intensity Values: {intensity_values}')

```
